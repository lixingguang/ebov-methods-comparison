```python
%%R
library(ape)
library(lubridate)
library(magrittr)
```

```python
%%R
s <- read.dna("data/Makona_1610_cds_ig.fas",format="fasta",as.matrix=TRUE)
tipnames <- row.names(s)
tipdates <- strsplit(tipnames,"_") %>% lapply(.,tail,1) %>% unlist %>% ymd %>% decimal_date
df <- data.frame(node_name=tipnames,date=tipdates)
write.table(df,"trees/Makona_1610_cds_ig.dates",col.names=TRUE,row.names=FALSE,quote=FALSE,sep=",")
df <- data.frame(c(length(tipnames),tipnames),c("",tipdates))
write.table(df,"trees/Makona_1610_cds_ig.td",col.names=FALSE,row.names=FALSE,quote=FALSE,sep="\t")
```
